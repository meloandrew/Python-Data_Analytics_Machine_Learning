## Example

Find the minimum value for the function y = (x + 5)Â², consider x = 3 (initial value)

### Using Descent Gradient

```python
# Weight Initial Value:
cur_x = 3 

# Learning rate:
rate = 0.01 

# Stop criteria for the algorithm:
precision = 0.000001 
max_iters = 10000

# Previous step size:
previous_step_size = 1 

# Iteration counter:
iters = 0 

# Function Gradient:
gf = lambda x: 2*(x+5) 
```

```python
while previous_step_size > precision and iters < max_iters:
    
    # Store weight current value:
    prev_x = cur_x 
    
    # Use Descent Gradient:
    cur_x = cur_x - rate * gf(prev_x) 
    
    # Update weight value:
    previous_step_size = abs(cur_x - prev_x) 
    
    # Increment iteration value:
    iters = iters + 1 
    
    # Print iterations:
    print("Iteration", iters,"\nWeight value is equal to ", cur_x) 
    
print("\nFunction Minimum value: ", cur_x)
```
